```python
from llama_cpp import Llama, LlamaLogitsProcessor, LlamaLogitsProcessorList
import numpy as np

class LlamaAutocompleteProcessor(LlamaLogitsProcessor):
    def __init__(self, llama_instance, selection_func):
        self.llm = llama_instance
        self.selection_func = selection_func

    def __call__(self, input_ids, scores):
        """
        input_ids: List of token IDs generated so far
        scores: Numpy array of logits (shape: [vocab_size])
        """
        # 1. Decode the input tokens to text string equivalents
        decoded_text = self.llm.detokenize(input_ids).decode('utf-8')
        
        # 2. Run your original selection function logic
        selected_idx = self.selection_func([decoded_text])
        
        # 3. Mask out all unselected tokens in the numpy array
        # llama.cpp uses -inf or large negative numbers to mask
        mask = np.ones_like(scores, dtype=bool)
        mask[selected_idx] = False
        scores[mask] = -10000.0
        
        return scores
```
